```python
df_movie = pd.read_csv("IMDb Movies India.csv.zip", encoding="latin1")
df_movie.head()
```

```python
df_movie.dropna(subset=["Rating"], inplace=True)
df_movie.isnull().sum()
```

```python
df_movie.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre'],inplace=True)
df_movie.isnull().sum()
```

```python
df_movie.head()
```

```python
df_movie['Votes']=df_movie['Votes'].str.replace(',','').astype(int)
```

```python
df_movie['Year']=df_movie['Year'].str.strip('()').astype(int)
```

```python
df_movie['Duration']=df_movie['Duration'].str.strip('min')
```

```python
df_movie['Duration'].fillna(df_movie['Duration'].median(),inplace=True)
```

```python
df_movie.isnull().sum()
```

```python
df_movie.info()
```

```python
df_movie.head()
```

```python
top_movies=df_movie.loc[df_movie['Rating'].sort_values(ascending=False)[:10].index]
top_movies
```

```python
sns.histplot(data=top_movies, x="Year",hue="Rating", multiple="stack")
plt.title('Distribution of Top 10 Movies', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Rating', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
genre_counts = df_movie['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

top_n_genres = genre_counts.head(5)
top_n_genres
```

```python
plt.figure(figsize=(4,4))
plt.pie(top_n_genres['Count'], labels=top_n_genres['Genre'], autopct='%.1f%%', startangle=140, colors=sns.color_palette('pastel')[0:5])
plt.title('Movie Genres Distribution',fontsize=15)
plt.axis('equal')
plt.show()
```

```python
director_avg_rating = df_movie.groupby('Director')['Rating'].mean().reset_index()
director_avg_rating = director_avg_rating.sort_values(by='Rating', ascending=False)
top_directors = director_avg_rating.head(10)
top_directors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_directors, x='Rating', y='Director', palette='viridis')
plt.title('Top 10 Directors by Average Rating',fontsize=15)
plt.xlabel('Average Rating',fontsize=15 )
plt.ylabel('Director',fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_movie, x='Rating', y='Votes')
plt.title('Votes vs. Rating',fontsize=30)
plt.xlabel('Rating',fontsize=15)
plt.ylabel('Votes',fontsize=15)
```

```python
actor_counts = df_movie['Actor 1'].value_counts().reset_index()
actor_counts.columns = ['Actor', 'MovieCount']
top_n_actors = actor_counts.head(10)
top_n_actors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_n_actors, x='MovieCount', y='Actor', orient='h')
plt.title('Top Actors by Number of Movie',fontsize=30)
plt.xlabel('Number of Movie',fontsize=25)
plt.ylabel('Actor',fontsize=30)
plt.show()
```

```python
yearly_movie_counts = df_movie['Year'].value_counts().reset_index()
yearly_movie_counts.columns = ['Year','MovieCount']
yearly_movie_counts = yearly_movie_counts.sort_values(by='Year')
yearly_movie_counts
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_movie_counts, x='Year', y='MovieCount')
plt.title('Number of Movies Released Every Year',fontsize=30)
plt.xlabel('Year',fontsize=24)
plt.ylabel('Number of Movies Released',fontsize=30)
plt.show()
```

```python
filtered_df = df_movie[(df_movie['Rating'] >= 8) & (df_movie['Votes']>10000)]
filtered_df.head(10)
```

```python
plt.figure(figsize=(12, 6))
ax=sns.barplot(data=filtered_df, x='Name', y='Votes', hue='Rating',dodge=False,width=0.5,palette='muted')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
ax.legend(loc='upper right')
ax.set_xticklabels('Movie Name')
ax.set_yticklabels('Votes')
ax.set_title('Movies with rating greater than 8 and Votes greater than 10000')
plt.show()
```

```python
df_movie['Duration']=df_movie['Duration'].astype(int)
df_movie['Year']=df_movie['Year'].astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_movie, x='Year', y='Duration', errorbar=None)
plt.xlabel('Year')
plt.ylabel('Duration in  minutes')
plt.title('Duartion of movie by year')
plt.xticks(np.arange(1990,2024, 5))
plt.show()
```

```python
df_movie['Genre'] = df_movie['Genre'].str.split(',')
genre_df = df_movie.explode('Genre')
genre_df
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=genre_df, x='Genre', order=genre_df['Genre'].value_counts().index,palette='viridis')
plt.xlabel('Number of Movies',fontsize=14)
plt.ylabel('Genre',fontsize=14)
plt.title('Number of Movies for Each Genre',fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
average_rating_by_genre = genre_df.groupby('Genre')['Rating'].mean().reset_index()
average_rating_by_genre + average_rating_by_genre.sort_values(by='Rating', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=average_rating_by_genre, y='Rating', x='Genre', palette='coolwarm')
plt.ylabel('Average Rating',fontsize=14)
plt.xlabel('Genre',fontsize=14)
plt.title('Average rating of movies in each genre',fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
trans_data = df_movie.drop(['Name'],axis=1)
trans_data['Director'] = labelencoder.fit_transform(df_movie['Director'])
trans_data['Actor 1'] = labelencoder.fit_transform(df_movie['Actor 1'])
trans_data['Actor 2'] = labelencoder.fit_transform(df_movie['Actor 2'])
trans_data['Actor 3'] = labelencoder.fit_transform(df_movie['Actor 3'])
trans_data['Genre'] = labelencoder.fit_transform(df_movie['Genre'].apply(lambda x:','.join(x)))
trans_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
sc_data = scaler.fit_transform(trans_data)
sc_df = pd.DataFrame(sc_data, columns=trans_data.columns)
sc_df.head()
```

```python
corr_df = trans_data.corr(numeric_only=True)
corr_df['Rating'].sort_values(ascending=False)
```

```python
sns.heatmap(corr_df, annot=False, cmap='coolwarm')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
```

```python
x = trans_data.drop('Rating', axis=1)
y= trans_data['Rating']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
x_train = np.array(x_train)
y_pred = model.predict(x_train)
y_pred
```

```python
print('R2 score: ',r2_score(y_train, y_pred))
print('Mean squared Error:',mean_squared_error(y_train, y_pred))
print('Mean Absolute Error: ',mean_absolute_error(y_train, y_pred))
```
